```python
df=pd.read_csv("https://raw.githubusercontent.com/Premalatha-success/Datasets/main/titanic-training-data.csv")
```

```python
sns.relplot(x="Sex",y="PassengerId",data=df)
```

```python
sns.catplot(x="Age",y="Fare",kind="box",data=df)
```

```python
sns.catplot(x="Age",y="Fare",hue="Sex",kind="violin",data=df)
```

```python
sns.catplot(x="Age",y="Fare",hue="Sex",kind="violin",split=True,data=df)
```

```python
sns.catplot(x="Age",y="Fare",hue="Sex",kind="bar",data=df)
```

```python
sns.catplot(x="Age",y="Fare",hue="Sex",kind="point",data=df)
```

```python
sns.catplot(x="Age",y="Fare",hue="Sex",aspect=.9,kind="swarm",data=df)
```
